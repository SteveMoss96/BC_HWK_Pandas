## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate some totals, averages, and percentage metrics. Use 'len' and 'unique' to calc length in the Series
total_schools = len(school_data_complete_df["School ID"].unique())

# Calculate the total number of students, as above
total_students = len(school_data_complete_df["Student ID"].unique())

# Calculate the total budget, use '.sum' to calc total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score, use '.mean' to calc average
avg_math_score = school_data_complete_df["math_score"].mean()

# Calculate the average reading score, a
avg_reading_score = school_data_complete_df["reading_score"].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
percent_passing_math = sum(school_data_complete_df["math_score"] >= 70) / total_students * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
percent_passing_reading = sum(school_data_complete_df["reading_score"] >= 70) / total_students * 100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing_rate = (percent_passing_math+percent_passing_reading)/2
# overall_passing_rate
```

```python
# Create a dataframe to hold the above District Summary results
District_Summary = pd.DataFrame({"Number of Schools":[total_schools],
                                "Number of Students":[total_students],
                                "Total Budget":[total_budget],
                                "Average Math score":[avg_math_score],
                                "Average Reading score":[avg_reading_score],
                                "Percentage Math passes":[percent_passing_math],
                                "Percentage Reading passes":[percent_passing_reading],
                                "Overall passing rate":[overall_passing_rate]
                                })
District_Summary.head()
```

```python
# Understand data types Index, Datatype and Memory information
District_Summary.info()
```

```python
# Give the District_Summary dataframe cleaner formatting
# EXAMPLE to use:  file_df["avg_cost"] = file_df["avg_cost"].map("${:.2f}".format)
District_Summary["Number of Schools"]=District_Summary["Number of Schools"].map("{:,}".format)
District_Summary["Number of Students"]=District_Summary["Number of Students"].map("{:,}".format)
District_Summary["Total Budget"]=District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary["Average Math score"]=District_Summary["Average Math score"].map("{:.2f}".format)
District_Summary["Average Reading score"]=District_Summary["Average Reading score"].map("{:.2f}".format)
District_Summary["Percentage Math passes"]=District_Summary["Percentage Math passes"].map("{:.2f}%".format)
District_Summary["Percentage Reading passes"]=District_Summary["Percentage Reading passes"].map("{:.2f}%".format)
District_Summary["Overall passing rate"]=District_Summary["Overall passing rate"].map("{:.2f}%".format)

District_Summary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group the data by School. Use:'df. groupby( col) - Return a groupby object for values from one column - use school name
school_name = school_data_complete_df.set_index('school_name').groupby(['school_name'])
school_name.head()
```

```python
# Check the information in the new group
school_name["Student ID"].count()
```

```python
# Find the school types
school_type=school_data.set_index(["school_name"])["type"]
# school_type  

# Find the number of Total Students at each school
school_total_students=school_name["student_name"].count()
# school_total_students

# Find the Total school budget for each school
school_budget=school_data_complete_df.groupby(["school_name"]).mean()["budget"]
# school_budget

#   Find the TPer student budget for each school
per_student_budget=school_budget/school_total_students
# per_student_budget

# Find the Average math score for each school 
school_av_math=school_name["math_score"].mean()
# school_av_math

# Find the Average reading score for each school   
school_av_reading=school_name["reading_score"].mean()
# school_av_reading

#Find the percent passing maths, score of >70, for each school
school_perc_passing_math=(school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].groupby("school_name").count()["student_name"]) / (school_total_students) * 100
# school_perc_passing_math

# #Find the percent passing reading, score of >70, for each school
school_perc_passing_reading=(school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].groupby("school_name").count()["student_name"]) / (school_total_students) * 100
# school_perc_passing_reading

# Find the overall passing rate for each school
overall_passing_rate = (school_data_complete_df[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)].groupby("school_name").count()["student_name"]) / (school_total_students) * 100
# overall_passing_rate
```

```python
# Create a dataframe to tabulate the School Summary results
School_Summary = pd.DataFrame({"School Type":school_type,
                                "Number of Students":school_total_students,
                                "School Budget":school_budget,
                                "Per student budget":per_student_budget,
                                "Average Math score":school_av_math,
                                "Average Reading score":school_av_reading,
                                "Percentage Math passes":school_perc_passing_math,
                                "Percentage Reading passes":school_perc_passing_reading,
                                "Overall passing rate":overall_passing_rate
                                })
School_Summary.head()
```

```python
# give the School Summary data cleaner formatting
# School_Summary["School Type"]=School_Summary["School Type"].map("{:,}".format)
School_Summary["Number of Students"]=School_Summary["Number of Students"].map("{:,}".format)
School_Summary["School Budget"]=School_Summary["School Budget"].map("${:,.2f}".format)
School_Summary["Per student budget"]=School_Summary["Per student budget"].map("${:,.2f}".format)
School_Summary["Average Math score"]=School_Summary["Average Math score"].map("{:.2f}".format)
School_Summary["Average Reading score"]=School_Summary["Average Reading score"].map("{:.2f}".format)
School_Summary["Percentage Math passes"]=School_Summary["Percentage Math passes"].map("{:.2f}%".format)
School_Summary["Percentage Reading passes"]=School_Summary["Percentage Reading passes"].map("{:.2f}%".format)
School_Summary["Overall passing rate"]=School_Summary["Overall passing rate"].map("{:.2f}%".format)

School_Summary.head()
```
